## Na√Øve Bayes Model Training

We set the new columns here, and when the new columns equal 0, means the review is negative, when the new columns equal 1, means the review is postive.

```python
# set new column
data_neg['@class']=0
data_pos['@class']=1
```

```python
data_neg=data_neg.drop(['sum'])
data_pos=data_pos.drop(['sum'])
```

```python
data_neg.index=data_neg.index+1000
```

```python
data_neg
```

```python
    # combine negative reviews and positve reviews together
result = data_pos.append(data_neg, sort=False)
```

```python
result
```

```python
# turn the class column to the last one for easier to see
df_id = result['@class']
result = result.drop('@class',axis=1)
result.insert(39354,'@class',df_id)  
```

```python
result
```

```python
result=result.fillna(0)
```

```python
X=result.drop(columns=['@class'])
```

```python
Y=result['@class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# finish the train test split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.25,random_state = 12)
```

Choosing MultinominalNB can solve most document classification problems

```python
# load multinomialNB from sklearn library
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB()
```

```python
clf.fit(X_train, Y_train)
```

```python
clf.score(X_test,Y_test)
```

## Model Testing

```python
predicted = clf.predict(X_test)
predicted
```

```python
probs = clf.predict_proba(X_test)
probs
```

## Extract most informative 30 features

```python
# use coef to make list
list_coef=clf.coef_[0]
```

```python
# turn the columns of X_train into list
column_clf=X_train.columns.to_list()
```

```python
# combine this two columns
data3=pd.DataFrame(
    {'feature': column_clf,
        'coef': list_coef,
     
     
    })
```

As shown in the above 30 most informative features, we can find that there are some positive features like 'good', 'like' and 'best' that have done effect to make the review tends to be positive.  

**Model Evaluation**

```python
# load metrics from sklearn library
from sklearn import metrics 
print(metrics.accuracy_score(Y_test, predicted))
print(metrics.roc_auc_score(Y_test, probs[:, 1]))
print(metrics.confusion_matrix(Y_test, predicted))
print(metrics.classification_report(Y_test, predicted))
```
